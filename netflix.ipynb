# Netflix Data Analysis
This project analyzes the distribution of Netflix content to understand trends in content production over the years.

*  ** Movies** are significantly more common than TV Shows.
* The most popular genre is **Documentary**


*   The country producing the most content is the **USA**
*   Content production peaked between **2016-2020**


*   A decline is observed after **2018**

Netflix content has shown a strong upward trend over the years, especially after 2016. The decline observed after 2018 may be due to dataset limitations or incomplete recent data.
